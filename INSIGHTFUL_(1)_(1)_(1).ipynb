```python
import string

PUNCT_TO_REMOVE = string.punctuation

def remove_punct(text):
   if isinstance(text, str):
     return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))
   else:
        return ""

df["article"] = df["article"].apply(remove_punct)
df["highlights"] = df["highlights"].apply(remove_punct)

df[["article", "highlights"]].head()
```

```python
import re

def contains_emoji(text):
    emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F"
    u"\U0001F300-\U0001F5FF"
    u"\U0001F680-\U0001F6FF"
    u"\U0001F1E0-\U0001F1FF"
    "]+", flags=re.UNICODE)
    return bool(emoji_pattern.search(text))

num_articles_with_emoji = df["article"].apply(contains_emoji).sum()
num_highlights_with_emoji = df["highlights"].apply(contains_emoji).sum()

print(f"Articles with emoji: {num_articles_with_emoji}")
print(f"Highlights with emoji: {num_highlights_with_emoji}")
```

```python
import torch
from datasets import load_dataset
from transformers import BartForConditionalGeneration, BartTokenizer, Trainer, TrainingArguments, DataCollatorForSeq2Seq
from rouge_score import rouge_scorer

# Define device globally
device = 0 if torch.cuda.is_available() else -1
torch_device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python

train_data = dataset["train"]
val_data = dataset["validation"]
test_data = dataset["test"]

print(f"Train size: {len(train_data)} | Val size: {len(val_data)} | Test size: {len(test_data)}")
```

```python
from transformers import BartTokenizer

# اسم النموذج
model_name = "facebook/bart-large-cnn"

# تحميل التوكنيزر
tokenizer = BartTokenizer.from_pretrained(model_name)

# ↓↓↓ ضعها هنا ↓↓↓
train_data = train_data.select(range(50000))
val_data = val_data.select(range(7000))
# ↑↑↑ ضعها هنا ↑↑↑

# أطوال الإدخال والإخراج
max_input_length = 512
max_target_length = 128

# دالة المعالجة المسبقة
def preprocess(examples):
    inputs = examples["article"]          # المقالات
    targets = examples["highlights"]      # الملخصات

    model_inputs = tokenizer(
        inputs,
        max_length=max_input_length,
        truncation=True,
        padding="max_length"
    )

    labels = tokenizer(
        targets,
        max_length=max_target_length,
        truncation=True,
        padding="max_length"
    )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# تطبيق الدالة على بيانات التدريب والتحقق
tokenized_train = train_data.map(
    preprocess,
    batched=True,
    remove_columns=["article", "highlights", "id"]
)

tokenized_val = val_data.map(
    preprocess,
    batched=True,
    remove_columns=["article", "highlights", "id"]
)
```

```python
model = BartForConditionalGeneration.from_pretrained(model_name)
model.to(torch_device)
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
from transformers import Trainer, TrainingArguments, EarlyStoppingCallback
import torch
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="steps",
    eval_steps=3000,
    save_strategy="steps",        # 🔹 must match eval_strategy
    save_steps=3000,              # 🔹 save every 1000 steps
    logging_steps=3000,
    learning_rate=3e-5,
    warmup_steps=500,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=2,
    num_train_epochs=1,
    weight_decay=0.01,
    save_total_limit=1,
    logging_dir="./logs",
    fp16=torch.cuda.is_available(),
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    report_to="none"
)


# Trainer مع Early Stopping
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train.select(range(20000)),
    eval_dataset=tokenized_val.select(range(2000)),
    tokenizer=tokenizer,
    data_collator=data_collator,
) # Added missing parenthesis here


trainer.train()
```

```python
from rouge_score import rouge_scorer
import numpy as np

# نفس الـ scorer
scorer = rouge_scorer.RougeScorer(['rouge1','rouge2','rougeL'], use_stemmer=True)

def evaluate_summaries(model, tokenizer, dataset, n_samples=50):
    model.eval()
    scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}
    for i in range(n_samples):
        article = dataset[i]["article"]
        reference = dataset[i]["highlights"]
        # تجهيز الإدخال
        inputs = tokenizer([article], return_tensors="pt", truncation=True, max_length=512).to(model.device)
        # توليد الملخص
        with torch.no_grad():
            summary_ids = model.generate(**inputs, max_length=128, num_beams=4, early_stopping=True)
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        score = scorer.score(reference, summary)
        # إضافة القيم إلى القوائم
        for k in scores.keys():
            scores[k].append(score[k].fmeasure)

    # حساب المتوسطات
    avg_scores = {k: np.mean(v) for k, v in scores.items()}
    return avg_scores

# استدعاء التقييم
rouge_scores = evaluate_summaries(model, tokenizer, val_data, n_samples=50)

# طباعة النتائج
print("Average ROUGE scores:")
for k, v in rouge_scores.items():
    print(f"{k}: {v:.4f}")
```

```python
import numpy
import torch
from transformers import pipeline, MarianTokenizer, MarianMTModel

device = 0 if torch.cuda.is_available() else -1
torch_device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# -----------------------------
# ✅ Assume you already have:
# model -> your trained BART model
# tokenizer -> corresponding tokenizer
# -----------------------------

print("✅ Using your trained BART model and tokenizer for summarization...")

print("✅ Loading sentiment model...")
sentiment_model = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-roberta-base-sentiment",
    device=device
)

print("✅ Loading translation models...")
# English → Arabic
en2ar_tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-ar")
en2ar_model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-ar")
# Arabic → English
ar2en_tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-ar-en")
ar2en_model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-ar-en")

print("✅ All models ready!")

# -----------------------------
# 🔹 Translation function
# -----------------------------
def translate_hf(text, target_lang):
    if target_lang == "ar":
        tokenizer_hf, model_hf = en2ar_tokenizer, en2ar_model
    else:
        tokenizer_hf, model_hf = ar2en_tokenizer, ar2en_model
    inputs = tokenizer_hf(text, return_tensors="pt", truncation=True)
    outputs = model_hf.generate(**inputs, max_length=200)
    return tokenizer_hf.decode(outputs[0], skip_special_tokens=True)

# -----------------------------
# 🔹 Master function
# -----------------------------
def process_text_master(text):
    print("\n===============================")
    print("✅ Step 1 — Sentiment Analysis")
    print("===============================\n")

    sentiment_raw = sentiment_model(text[:400])[0]
    label_map = {"LABEL_0": ("Negative", "☹️"), "LABEL_1": ("Neutral", "😐"), "LABEL_2": ("Positive", "😊")}
    fixed_label, emoji = label_map.get(sentiment_raw["label"], ("Unknown", ""))
    sentiment = {"label": fixed_label, "score": sentiment_raw["score"]}
    print(f"📝 Sentiment: {fixed_label} {emoji} ({sentiment_raw['score']:.2f})")

    print("\n===============================")
    print("✅ Step 2 — Auto Content Type Detection")
    print("===============================\n")

    t = text.lower()
    if any(w in t for w in ["breaking", "report", "news"]):
        detected, style_prompt = "News", "Summarize this news article focusing on clarity and key facts:"
    elif any(w in t for w in ["study", "experiment", "data", "results"]):
        detected, style_prompt = "Research", "Summarize this research text focusing on findings:"
    elif any(w in t for w in ["😭", "😂", "💙", "#", "followers", "instagram", "twitter"]):
        detected, style_prompt = "Social Media", "Summarize this social media post in an emotional friendly tone:"
    elif any(w in t for w in ["story", "once upon", "adventure", "journey"]):
        detected, style_prompt = "Story", "Summarize this story narratively keeping main events:"
    else:
        detected, style_prompt = "General", "Summarize this text clearly and concisely:"
    print(f"✅ Detected Content Type: {detected}")

    print("\n===============================")
    print("✅ Step 3 — Smart Summarization")
    print("===============================\n")

    # 🔹 Use your trained BART model and tokenizer
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512).to(torch_device)

    summary_ids = model.generate(
        inputs["input_ids"],
        max_length=150,
        num_beams=4,
        min_length=20,
        length_penalty=1.2,
        early_stopping=True
    )


    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)

    print("📃 Summary:")
    print(summary)

    print("\n===============================")
    print("✅ Step 4 — Translation Options")
    print("===============================\n")

    lang_choice = input("🌍 Translate summary to: 1) Arabic  2) English  3) No Translation → ").strip()
    translated = None
    if lang_choice == "1":
        translated = translate_hf(summary, "ar")
    elif lang_choice == "2":
        translated = translate_hf(summary, "en")

    if translated:
        print("🈂️ Translated Summary:", translated)
    else:
        print("✅ No translation selected.")

    return {"sentiment": sentiment, "content_type": detected, "summary": summary, "translated": translated}

print("\n✅ MASTER FUNCTION READY!")
```

```python
# ================================
# ✅ CELL 1 — Load Qwen Model once
# ================================

from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

print("Loading Qwen 1.5B Instruct...")

reasoning_model_name = "Qwen/Qwen2.5-1.5B-Instruct"

qwen_tokenizer = AutoTokenizer.from_pretrained(reasoning_model_name)
qwen_model = AutoModelForCausalLM.from_pretrained(reasoning_model_name)

# ✅ This is the global pipeline you will reuse everywhere
qa_pipe = pipeline(
    "text-generation",
    model=qwen_model,
    tokenizer=qwen_tokenizer,
    max_new_tokens=150,
    temperature=0.3
)

print("✅ Qwen is loaded and ready to use!")
```

```python
# ============================================================
# ✅ CELL: Full-text Q&A using Qwen (qa_pipe already loaded)
# ============================================================

def smart_answer(full_text, question):
    # Truncate full_text to avoid OutOfMemoryError for very long inputs
    context_text = full_text[:2000] # Limiting context to 2000 characters
    prompt = f"""
You are a smart reasoning assistant.
Do NOT copy or repeat any sentence from the text.

TEXT (context only):
"""{context_text}"""

QUESTION:
{question}

GUIDELINES:
- NEVER copy phrasing from the text.
- ALWAYS answer in your OWN words.
- Be short, clear, and original.
- If the answer is in the text, rephrase it in your style.
- If it is not in the text, give a smart general explanation.

ANSWER:
"""
    out = qa_pipe(prompt)[0]["generated_text"]
    answer = out.split("ANSWER:")[-1].strip()
    return answer


# ✅ Q&A loop using smart_answer (ASK multiple questions until 'done')
def chat_over_text(full_text):
    print("\n💬 Q&A mode started. Type 'done' to exit.\n")

    while True:
        question = input("❓ Your question: ").strip()

        if question.lower() in ["done", "stop", "exit", "quit", "خلص", "انهاء"]:
            print("\n✅ Q&A mode ended.")
            break

        answer = smart_answer(full_text, question)
        print("\n💬 Answer:", answer, "\n")
```

```python
# ================================
# ✅ Movie Summarizer using preloaded BART + Qwen + OMDb
# ================================

import requests, json
import torch

# الموديلات محملة بالفعل
# model, tokenizer, qa_pipe, torch_device موجودة

def summarize_text_bart(text, model=model, tokenizer=tokenizer, max_input_length=512, max_summary_length=120):
    if not isinstance(text, str) or len(text.strip()) == 0:
        return ""

    inputs = tokenizer([text], return_tensors="pt", truncation=True, max_length=max_input_length, padding="longest")
    inputs = {k: v.to(torch_device) for k, v in inputs.items()}

    summary_ids = model.generate(
        inputs["input_ids"],
        num_beams=4,
        max_length=max_summary_length,
        min_length=20,
        length_penalty=1.2,
        early_stopping=True
    )
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary.strip()

def extract_theme_from_summary(summary):
    prompt = f"""
You are a professional movie analyst.
Explain the movie's main theme in ONE original intelligent sentence.
Do NOT copy any words from the summary.

SUMMARY:
{summary}

THEME:
"""
    out = qa_pipe(prompt)[0]["generated_text"]
    theme = out.split("THEME:")[-1].strip()
    return theme

OMDB_KEY = "ff0efb1e"

def fetch_movie_from_omdb(title):
    url = f"http://www.omdbapi.com/?t={requests.utils.quote(title)}&plot=full&apikey={OMDB_KEY}"
    data = requests.get(url).json()
    if not data or data.get("Response") == "False":
        return None, None
    return data, data.get("Plot", "")

def summarize_movie(movie_name):
    movie_data, plot = fetch_movie_from_omdb(movie_name)
    if movie_data is None:
        return {"error": f"Movie '{movie_name}' not found."}, None

    info = {
        "title": movie_data.get("Title", ""),
        "genre": movie_data.get("Genre", ""),
        "runtime": movie_data.get("Runtime", ""),
        "poster": movie_data.get("Poster", ""),
        "actors_full": movie_data.get("Actors", ""),
        "main_actor": movie_data.get("Actors", "").split(",")[0].strip() if movie_data.get("Actors") else "",
    }

    summary = summarize_text_bart(plot)  # يستخدم الموديل المحمّل بالفعل
    info["summary"] = summary

    theme = extract_theme_from_summary(summary)  # يستخدم qa_pipe
    info["theme"] = theme

    return info, plot

# ================================
# ✅ Run Movie Summarizer
# ================================
while True:
    movie_name = input("\n🎬 Enter movie title (or type 'done' to exit): ").strip()
    if movie_name.lower() in ["done", "exit", "quit", "no"]:
        print("\n✅ Program ended.")
        break

    info, plot_text = summarize_movie(movie_name)
    print(json.dumps(info, indent=2, ensure_ascii=False))

    if plot_text:
        result = process_text_master(plot_text)
        print("\n🧠 TEXT PROCESSING RESULT:")
        print(result)

        chat_over_text(plot_text)
    else:
        print("\n❌ Could not retrieve plot for Q&A.")
```
